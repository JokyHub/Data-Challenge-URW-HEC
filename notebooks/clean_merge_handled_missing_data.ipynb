# json mall name


```python
import json

# Load the JSON data from file
with open('mapping_mall.json', 'r') as file:
    mall_ids = json.load(file)
```

```python
mall_data['mall_id'] = mall_data['mall_name'].map(mall_ids)
mall_data
```

# merge data

```python
# Let's start by merging mall and store data using the mall_id as the key
mall_store_data = pd.merge(mall_data, store_data, on='mall_id', how='inner')
mall_store_data
```

```python
# Let's add the cost data of all stores to the mall_store_data
mall_store_cost_data = pd.merge(mall_store_data, cost_data, on='store_id', how='inner')
mall_store_cost_data
```

```python
dwell_traffic_data = pd.merge(dwell_time_data, referenciel_traffic_data, on=['block_id','external_mall_id'], how='inner')
# dwell_traffic_data = pd.merge(dwell_time_data, referenciel_traffic_data, on='block_id', how='left')

dwell_traffic_data
```

```python
dwell_traffic_people_flow_data = pd.merge(dwell_traffic_data, people_flow_data, on=['block_id','external_mall_id'], how='inner')

dwell_traffic_people_flow_data
```

```python
dwell_traffic_people_flow_data.to_csv('traffic_data.csv')
```

```python
mall_store_cost_data.to_csv('store_cost_data.csv')
```

```python
all_data = pd.merge(dwell_traffic_people_flow_data, mall_store_cost_data, on=['store_id'], how='left')
all_data
```

# handle the missing data

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Display the first few rows to understand the data structure
first_rows = all_data.head()

# Print the summary of missing values per column
missing_info = all_data.isnull().sum()

# Visualize missing data with a heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(all_data.isnull(), cbar=False)
plt.title('Heatmap of Missing Data')
plt.show()
```

```python
# Visualize the percentage of missing data for each column
missing_percentage = (all_data.isnull().sum() / len(all_data)) * 100
plt.figure(figsize=(10, 5))
missing_percentage.plot(kind='bar', color='blue')
plt.title('Percentage of Missing Data by Column')
plt.ylabel('Percentage Missing')
plt.show()
```

```python
# Check if the pattern of missingness correlates with other columns
missing_corr = all_data.isnull().corr()  # Correlation of missingness between columns
plt.figure(figsize=(8, 6))
sns.heatmap(missing_corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Missing Data Patterns')
plt.show()
```

```python
# Fill missing values in 'store_year_closed' with '1/1/2099'
all_data['store_year_closed'] = all_data['store_year_closed'].fillna(pd.Timestamp('2099-01-01'))
```

```python
null_percentage = all_data.isnull().mean() * 100
null_percentage
```

```python
# Drop the 24 rows with no store and mall information
all_data.dropna(subset='mall_id', how='all', inplace=True)

null_percentage = all_data.isnull().mean() * 100
null_percentage
```

```python
from sklearn.impute import SimpleImputer

# drop columsn with over 75% missing
all_data.drop(columns=['rent_incentive_r12m', 'administration_fees_r12m', 'large_scale_works_r12m'], inplace=True)

# sales and rent are import columns, filling those with imputed data may confuse the analysis, so leave them blank
# median to fill the rest
median_imputer = SimpleImputer(strategy='median')
all_data[['service_charges_r12m', 'real_estate_taxes_r12m', 'maintenance_works_r12m', 'marketing_costs_r12m']] = median_imputer.fit_transform(all_data[['service_charges_r12m', 'real_estate_taxes_r12m', 'maintenance_works_r12m','marketing_costs_r12m']])
```

```python
null_percentage = all_data.isnull().mean() * 100
null_percentage
```
