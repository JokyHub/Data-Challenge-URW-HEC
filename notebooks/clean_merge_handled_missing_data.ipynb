```python
# find out which columns are most related with the columns with missing data

from sklearn.preprocessing import LabelEncoder

df = all_data.copy()

datetime_cols = df.select_dtypes(include=['datetime', 'datetime64']).columns
df = df.drop(columns=datetime_cols)

# Convert categorical string columns to numbers
# Detect object type columns that need conversion
categorical_cols = df.select_dtypes(include=['object']).columns
label_encoders = {}
for col in categorical_cols:
    le = LabelEncoder()
    # Fitting label encoder to each categorical column
    df[col] = le.fit_transform(df[col].astype(str))
    label_encoders[col] = le  # Store label encoder for potential inverse transform

target_columns = ['service_charges_r12m', 'marketing_costs_r12m', 'real_estate_taxes_r12m', 'maintenance_works_r12m']
correlation_matrix = df.corr()
target_correlation = correlation_matrix[target_columns].drop(index=target_columns)

strongest_relations = {}
for column in target_columns:
    strongest_relations[column] = target_correlation[column].abs().idxmax()

strongest_relations
```

```python
# Define the strongest relationships for filling missing values
fill_rules = {
    'service_charges_r12m': 'mall_country',
    'marketing_costs_r12m': 'store_visits_monthly_average',
    'real_estate_taxes_r12m': 'external_mall_id',
    'maintenance_works_r12m': 'mall_id'
}

# Ensure the grouping columns are treated as categorical if necessary
categorical_cols = ['mall_country', 'external_mall_id', 'mall_id']
for col in categorical_cols:
    if col in df.columns:
        df[col] = df[col].astype(str)  # Ensure consistent grouping

# Fill missing values
for target_col, related_col in fill_rules.items():
    if target_col in df.columns and related_col in df.columns:
        if target_col == 'marketing_costs_r12m':
            # Use linear regression or similar method to predict marketing_costs_r12m based on store_visits_monthly_average
            from sklearn.linear_model import LinearRegression
            model = LinearRegression()
            not_null_data = df[df[target_col].notna() & df[related_col].notna()]
            model.fit(not_null_data[[related_col]], not_null_data[target_col])
            null_data = df[df[target_col].isna() & df[related_col].notna()]
            predicted_values = model.predict(null_data[[related_col]])
            df.loc[df[target_col].isna(), target_col] = predicted_values
        else:
            # Group by the related column and calculate the median
            median_values = df.groupby(related_col)[target_col].median()
            missing_mask = df[target_col].isna()
            df.loc[missing_mask, target_col] = df.loc[missing_mask, related_col].map(median_values)
```

```python
# Update the original DataFrame with the filled values
all_data.update(df[fill_rules.keys()])

null_percentage = all_data.isnull().mean() * 100
null_percentage
```
