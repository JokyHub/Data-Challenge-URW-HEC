# merge data

```python
# Let's start by merging mall and store data using the mall_id as the key
mall_store_data = pd.merge(mall_data, store_data, on='mall_id', how='inner')
mall_store_data
```

```python
# Let's add the cost data of all stores to the mall_store_data
mall_store_cost_data = pd.merge(mall_store_data, cost_data, on='store_id', how='inner')
mall_store_cost_data
```

```python
dwell_traffic_data = pd.merge(dwell_time_data, referenciel_traffic_data, on=['block_id','external_mall_id'], how='inner')
# dwell_traffic_data = pd.merge(dwell_time_data, referenciel_traffic_data, on='block_id', how='left')

dwell_traffic_data
```

```python
dwell_traffic_people_flow_data = pd.merge(dwell_traffic_data, people_flow_data, on=['block_id','external_mall_id'], how='inner')

dwell_traffic_people_flow_data
```

```python
dwell_traffic_people_flow_data.to_csv('traffic_data.csv')
```

```python
mall_store_cost_data.to_csv('store_cost_data.csv')
```

```python
print(len(set(mall_store_cost_data['store_id'])))
print(len(set(dwell_traffic_people_flow_data['store_id'])))
```

```python
all_data = pd.merge(dwell_traffic_people_flow_data, mall_store_cost_data, on=['store_id'], how='left')
all_data
```
