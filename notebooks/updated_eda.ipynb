## time features

convert the store open and close date to datetime, get the number of days the store lived until today

```python
data['store_year_opened'] = pd.to_datetime(data['store_year_opened'], format='%d/%m/%Y')
data['store_year_closed'] = pd.to_datetime(data['store_year_closed'], format='%d/%m/%Y', errors='coerce')  # Coerce errors for any invalid or missing entries
```

```python
from datetime import datetime

today = pd.to_datetime(datetime.now().strftime('%d/%m/%Y'))

data['living_days'] = data.apply(
    lambda row: (row['store_year_closed'] - row['store_year_opened']).days if pd.notnull(row['store_year_closed']) and row['store_year_closed']< pd.Timestamp('2099-01-01') else (today - row['store_year_opened']).days,
    axis=1
)
```

## Cost Analysis

```python
# show some example cost percentages
store_ids = [5697, 7179, 7288]
cost_labels = ['Rent', 'Service Charges', 'Marketing', 'Real Estate Taxes', 'Maintenance']

plt.figure(figsize=(12, 4))

for index, store_id in enumerate(store_ids):
    specific_store = data[data['store_id'] == store_id]

    # Extract cost values
    cost_values = [
        specific_store['store_rent_r12m'].iloc[0],
        specific_store['service_charges_r12m'].iloc[0],
        specific_store['marketing_costs_r12m'].iloc[0],
        specific_store['real_estate_taxes_r12m'].iloc[0],
        specific_store['maintenance_works_r12m'].iloc[0]
    ]

    # Create a pie chart for each store
    plt.subplot(1, 3, index + 1)
    plt.pie(cost_values, labels=cost_labels, autopct='%1.1f%%')
    plt.title(f'Cost Breakdown for Store ID {store_id}')

plt.tight_layout()
plt.show()
```

```python
columns_of_interest = ['store_rent_r12m', 'service_charges_r12m',
    'marketing_costs_r12m', 'real_estate_taxes_r12m', 'maintenance_works_r12m'
]

average_values = data[columns_of_interest].mean()
median_values = data[columns_of_interest].median()

statistics_df = pd.DataFrame({
    'Average': average_values,
    'Median': median_values
})

fig, ax = plt.subplots(figsize=(10, 5))

avg_bars = statistics_df['Average'].plot(kind='bar', ax=ax, color='skyblue', position=1, width=0.4, label='Average')
med_bars = statistics_df['Median'].plot(kind='bar', ax=ax, color='orange', position=0, width=0.4, label='Median')

statistics_df['Average'].plot(kind='bar', ax=ax, color='skyblue', position=1, width=0.4, label='Average')
statistics_df['Median'].plot(kind='bar', ax=ax, color='orange', position=0, width=0.4, label='Median')

ax.set_title('Average and Median of Costs')
ax.set_ylabel('Values')
ax.set_xlabel('Categories')
ax.set_xticklabels(statistics_df.index, rotation=0)
ax.legend()

for bar in avg_bars.patches + med_bars.patches:
    value_format = f'{bar.get_height():,.0f}' if bar.get_height() > 10 else f'{bar.get_height():,.2f}'
    ax.annotate(value_format,
                (bar.get_x() + bar.get_width() / 2, bar.get_height()),
                ha='center', va='bottom',
                xytext=(0, 5),
                textcoords='offset points')

plt.tight_layout()
plt.show()
```

## Analyzing the Top Categories in Branch Levels

```python
plt.figure(figsize=(12, 18))
pivot_data = data_for_visiting.pivot_table(values='store_visits_daily_average', index='store_branch_level_2', aggfunc='mean')
sns.heatmap(pivot_data, cmap='coolwarm', annot=True)
plt.title('Heat Map of Daily Visitor Average by Branch Level 2')
plt.show()
```
