## time features

convert the store open and close date to datetime, get the number of days the store lived until today

```python
store_cost_data['store_year_opened'] = pd.to_datetime(store_cost_data['store_year_opened'], format='%d/%m/%Y')
store_cost_data['store_year_closed'] = pd.to_datetime(store_cost_data['store_year_closed'], format='%d/%m/%Y', errors='coerce')  # Coerce errors for any invalid or missing entries
```

```python
from datetime import datetime

today = pd.to_datetime(datetime.now().strftime('%d/%m/%Y'))

store_cost_data['living_days'] = store_cost_data.apply(
    lambda row: (row['store_year_closed'] - row['store_year_opened']).days if pd.notnull(row['store_year_closed']) else (today - row['store_year_opened']).days,
    axis=1
)
```

```python
store_cost_data
```

## Encode Text Features (label encode)

```python
from sklearn.preprocessing import LabelEncoder

le_mall_country = LabelEncoder()
le_store_branch_level_1 = LabelEncoder()
le_store_branch_level_2 = LabelEncoder()
le_store_branch_level_3 = LabelEncoder()


store_cost_data['mall_country'] = le_mall_country.fit_transform(store_cost_data['mall_country'])
store_cost_data['store_branch_level_1'] = le_store_branch_level_1.fit_transform(store_cost_data['store_branch_level_1'])
store_cost_data['store_branch_level_2'] = le_store_branch_level_2.fit_transform(store_cost_data['store_branch_level_2'])
store_cost_data['store_branch_level_3'] = le_store_branch_level_3.fit_transform(store_cost_data['store_branch_level_3'])

store_cost_data
```

## Total Cost

```python
store_cost_data['total_costs_r12m'] = (store_cost_data['store_rent_r12m'] +
                                  store_cost_data['service_charges_r12m'] +
                                  store_cost_data['marketing_costs_r12m'] +
                                  store_cost_data['administration_fees_r12m'] +
                                  store_cost_data['real_estate_taxes_r12m'] +
                                  store_cost_data['large_scale_works_r12m'] +
                                  store_cost_data['maintenance_works_r12m'])
```

## Ratios

```python
# total visit / window flow
traffic_data['visit_to_window_flow_ratio'] = traffic_data['store_total_visits'] / traffic_data['store_total_window_flow']

# sales / rent
store_cost_data['sales_to_rent_ratio'] = store_cost_data['store_sales_r12m'] / store_cost_data['store_rent_r12m']

# sales / cost
store_cost_data['sales_to_cost_ratio'] = store_cost_data['store_sales_r12m'] / store_cost_data['total_costs_r12m']
```
